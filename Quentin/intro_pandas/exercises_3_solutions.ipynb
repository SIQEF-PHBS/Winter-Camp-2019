```python
import pandas as pd
import numpy as np

stocks_filename = 'stocks_data.csv'
stocks_df = pd.read_csv(stocks_filename)
```

### Q1: Add two new column to `stocks_df` which contains the year and month for each row.

You can define a new column for a dataframe `df` using the following: `df['new_col_name'] = new_series`. Also, you can convert strings to numeric values using `pd.to_numeric`.

```python
stocks_df['year'] = pd.to_numeric(stocks_df.date.str[0:4])
stocks_df['month'] = pd.to_numeric(stocks_df.date.str[5:7])
stocks_df.head()
```

### Q2: What is the day with the maximum volume in one day for all stocks combined?

```python
stocks_df.groupby('date').volume.sum().max()
```

### Q3: What is the best and worst performer for the year 2010 using percentage change in `adj_close` as a metric?

```python
# Approach 1
grouped = stocks_df[stocks_df['year'] == 2010].groupby('ticker')
best_perf = (grouped.last()['adj_close'] / grouped.first()['adj_close']).idxmax()
worst_perf = (grouped.last()['adj_close'] / grouped.first()['adj_close']).idxmin()

print('Best performer: ' + str(best_perf))
print('Worst performer: ' + str(worst_perf))
```

### Q4: What are the best and worst performers for each year using `adj_close`?

```python
# Approach 1
best_worst_perf_df = pd.DataFrame(index=np.sort(stocks_df.year.unique()))

grouped = stocks_df.groupby(['ticker', 'year'])
annual_perf = (grouped.last().adj_close / grouped.first().adj_close - 1).reset_index(level=1).groupby('year')

best_worst_perf_df['best_ticker'] = annual_perf.idxmax() 
best_worst_perf_df['best_perf'] = annual_perf.max() 

best_worst_perf_df['worst_ticker'] = annual_perf.idxmin() 
best_worst_perf_df['worst_perf'] = annual_perf.min() 

best_worst_perf_df
```

```python
# Approach 2 
best_worst_perf_df = pd.DataFrame(index=np.sort(stocks_df.year.unique()))

grouped = stocks_df.groupby(['year', 'ticker'])
annual_perf = (grouped.last().adj_close / grouped.first().adj_close - 1).unstack()

def compute_best_worst(s):
    return s.idxmax(), s.max(), s.idxmin(), s.min()

annual_perf.apply(compute_best_worst, axis=1, result_type='expand')
```

### Q5: Compute the correlation of the percentage return of each stock with the percentage return of `AAPL` for the year 2016 using `adj_close`.

```python
def compute_corr(adj_close):
    return adj_close.pct_change().corr(aapl_pct_return)

stocks_df_date_idx = stocks_df[stocks_df.year == 2016].set_index('date')
aapl_pct_return = stocks_df_date_idx[stocks_df_date_idx.ticker == 'AAPL'].adj_close.pct_change()

return_corr = stocks_df_date_idx.groupby('ticker').adj_close.apply(compute_corr)
return_corr
```
