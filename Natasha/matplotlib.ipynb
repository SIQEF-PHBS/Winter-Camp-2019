# Python Plotting Basics

Python has several great options for creating nice looking plots

Today we will mostly be looking at **Matplotlib** -- the most popular plotting library in Python

* Based on MATLAB-style plotting
* Very flexible
* Not the most straight-forward syntax

Matplotlib has two interfaces:
1. MATLAB-style API
1. Object-oriented API

The MATLAB-style API is simpler to use, whilst the object-orientated API allows for greater control over the plot features

It is standard to import the library as follows

```python
import matplotlib.pyplot as plt
```

## MATLAB-style API

```python
import numpy as np


x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
plt.title('A sine function')
plt.plot(x, y, color='red', linewidth=4, linestyle=':')
plt.show()
```

```python
plt.plot(x, y, c='r', lw=4, ls=':')
plt.show()
```

```python
z = np.cos(x)**2

plt.plot(x, y, label='$f(x) = \sin(x)$', alpha=0.5)
plt.plot(x, z, label='$f(x) = \cos^2(x)$', alpha=0.5)
plt.legend()
plt.show()
```

Matplotlib also has methods for other types of plots, such as histograms, bar charts, scatter plots, etc.

The [Matplotlib gallery](https://matplotlib.org/gallery.html) is useful for finding examples of plots you might need

```python
data = np.random.uniform(size=100)
plt.hist(data, edgecolor='black', density=True)
plt.show()
```

## Object-oriented API

* The object-oriented API is particularly useful when you want to use subplots
* Here, the figure (the entire plot) and the axes (each subplot) are separate objects

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 3))

axes
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 3))

axes[0].plot(x, y)
plt.show()
```

```python
colors = ['red', 'blue', 'orange']

fig, axes = plt.subplots(1, 3, figsize=(15, 3))

for ax, color in zip(axes, colors):
    ax.plot(x, y, c=color, label=f'{color} plot')
    ax.legend()
    
plt.show()
```
